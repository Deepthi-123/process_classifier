```python
names = ['eventtime', 'msg_stamp','symbols' ,'eigen_force']
dataset = pandas.read_csv('/home/deepthi/task1_eig/label1_30.log', sep = ',', names = names)
lt = []
for i in dataset['msg_stamp'].map(str):
   lt.append('.' + i.zfill(3))
dataset['ros_time'] = pandas.Series(lt)
dataset['time_stamp'] = dataset['eventtime'].map(str) + dataset['ros_time']
del dataset['eventtime']
del dataset['msg_stamp']
del dataset['ros_time']
del dataset['symbols']
dataset_ = dataset.set_index('time_stamp')
eigen = pandas.Series(dataset_['eigen_force'])
```
